```python
#Import
import pandas as pd
import numpy as np
```

```python
#Create a list with columns
columns_main = ['Id','Name','Animal']
columns_sec = ['Id','Status_Register','Code','Name']
columns_third = ['Animal','Type_Animal','Code']

#Create dfs datas
data_main = [[423,'Blake','Dog'],
             [421,'Blake','Puppy'],
             [652,'Doki','Dog'],
             [652,'Nina','Dog'],
             [197,'Mia','Cat'],
             [816,'Luigi','Bird'],
             [782,'Pandora','Dog'],
             [340,'Xico','Cat'],
             [120,'Bolinha','Dog']]

data_sec = [[423,1,200,'Blake'],
            [421,1,150,'Blake'],
            [652,0,150,'Doki'],
            [652,0,200,'Nina'],
            [197,3,200,'Mia'],
            [816,2,150,'Luigi'],
            [782,1,150,'Pandora'],
            [340,2,200,'Xico'],
            [109,0,200,'Bolinha']]

data_third = [['Dog','Pet_0','A'],
              ['Cat','Pet_1','B'],
              ['Bird','Pet_2','C'],
              ['Fish','Pet_3','D']]
```

```python
#Create main df
df = pd.DataFrame(data_main, columns=columns_main)
df
```

```python
#Create df status
df_status = pd.DataFrame(data_sec, columns=columns_sec)
df_status
```

```python
#Create df type
df_type = pd.DataFrame(data_third, columns=columns_third)
df_type
```

# Merge

```python
#Inner merge by columns Id and Name
inner_merge = pd.merge(df, df_status[['Code','Id','Name']], how='inner', on=['Id','Name'])
inner_merge
```

```python
#Outer merge by column Id
outer_merge = pd.merge(df, df_status, how='outer', left_on=['Id'], right_on=['Id'], suffixes=('df', 'df_status'))
outer_merge
```

```python
#Left merge by dfs index
df_outer_merge = pd.merge(df, df_type, how='left', left_index=True, right_index=True)
df_outer_merge
```

```python
#Right merge by columns
right_merge = pd.merge(df, df_type, how='right')
right_merge
```

# Join

```python
#Inner join by index
df_join = df.join(df_status, how='inner', lsuffix='_left', rsuffix='_right')
df_join
```

```python
#Outer join by index and column Animal
df_join = df.join(df_type.set_index('Animal'), how='outer', on=['Animal'])
df_join
```

```python
#Left join by index and column Animal
df_join = df.join(df_type.set_index('Animal'), how='left', on=['Animal'], sort=True)
df_join
```

```python
#Right join by index
df_join = df.join(df_type, how='right', lsuffix='_left', rsuffix='_right')
df_join
```

# Concat

```python
#Concatenate rows
df_concat = pd.concat([df,df_status], axis=0)
df_concat
```

```python
#Concatenate columns
df_concat = pd.concat([df,df_status], keys=['df','df_status'], axis=1)
df_concat
```

```python
#Inner concatenate
df_concat = pd.concat([df,df_status], join='inner', ignore_index=True)
df_concat
```

```python
#Outer concatenate
df_concat = pd.concat([df,df_type], join='outer', ignore_index=True)
df_concat
```
