```python
import pandas as pd
import numpy as np
```

# Series

```python
#Create a Series with index by default
data = (8, 3, 0)

first_series = pd.Series(data)
first_series
```

```python
#Create a Series by passed index
data = (4, 7, 2)
index = ['A', 'B', 'D']

second_series = pd.Series(data, index=index)
second_series
```

```python
#Create a Series with duplicate index
data = (8, 3, 0)
index = ['A', 'A', 'B']

first_series = pd.Series(data, index=index)
first_series
```

```python
#Create a Series from a dictionary
dicionario = {'A':8, 'B':3, 'C':0}
first_series = pd.Series(dicionario)
first_series
```

```python
#Create a Series by passed index from a dictionary
dicionario = {'C':8, 'E':3, 'A':0}
index = ['A', 'B', 'C']
first_series = pd.Series(dicionario, index)
first_series
```

```python
#Get 'A' in Series
first_series.get('A')
```

```python
#iLoc in Series
first_series.iloc[2]
```

```python
#Std() Series
first_series.std()
```

```python
#Transform Series into an array
first_array = first_series.to_numpy()
first_array
```

```python
#Sum between Series
first_series = pd.Series((2,6,1), index=['A','B','C'])
second_series = pd.Series((4,3,8), index=['A','D','B'])

soma_series = first_series+second_series
soma_series
```

```python
#Create a Series with a name
first_series = pd.Series(data, name='first_name')
first_series
```

```python
#Rename Series
first_series = first_series.rename('changing_name')
first_series
```

```python
#Create a Dataframe from dicts of Series passing the index
df = pd.DataFrame(data_dict, index=['A','b','C'])
df
```

```python
#Create a Dataframe from dicts of Series passing the index and columns arguments
df = pd.DataFrame(data_dict, index=['A','b','C'], columns=['column_1','second_column'])
df
```
