```python
#Import lybrary and packages
import pandas as pd
import numpy as np
```

```python
#Create a dataframe with random numbers
df = pd.DataFrame(np.random.randint(0, 10, size=(100, 4)), columns=['Column1','Column2','Column3','Column4'])
df.head(4)
```

```python
#Df dimension
df.shape
```

```python
#Filter with loc: conditions
df.loc[(df.Column1 == 9) & (df.Column3 == 5)]
```

```python
#Filter with loc: select row range
df.loc[:6]
```

```python
#Filter with loc: select only a row
df.loc[[3]]
```

```python
#Filter with loc: select rows
df.loc[[1,4,10]]
```

```python
#Update with loc: update values
df.loc[df.Column1 == 9, 'Column2'] = 3
df.head()
```

iloc

```python
#Filter with iloc: select rows
df.iloc[[0,5,10,15]]
```

```python
#Filter with iloc: select only a row
df.iloc[[10]]
```

```python
#Filter with iloc: select only a column
df.iloc[:,[2]]
```

```python
#Filter with iloc: select only columns
df.iloc[:,[2,3]]
```

```python
#Filter with iloc: select rows and columns by range
df.iloc[10:13, 1:3]
```

```python
#Filter with iloc: select rows and columns by index
df.iloc[[7,15,32],[1,2,3]]
```

```python
#Filter with iloc: select rows by the range
df.iloc[2:3]
```

```python
#Filter with iloc: select columns by the range
df.iloc[:,1:2]
```

```python
#Filter with loc by rows range and by columns labels
df.loc[0: 4, ['Column1', 'Column2']]
```

```python
#Group data with groupby
df.loc [2:, ['Column2', 'Column3']].groupby('Column2').mean().reset_index()
```
